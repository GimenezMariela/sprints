1.2 Mostra per pantalla:

La quantitat total de cada recurs.
El codi d'estat de cada petició.

```python
URL = 'https://jsonplaceholder.typicode.com/'
respuesta = requests.get(URL)

recursos = ["posts", "comments", "albums", "photos", "todos", "users"]


for recurso in recursos:

    response = requests.get(URL + recurso)
    print(f"{recurso.upper()}: {len(response.json())} ")
    print(f" * Código de estado: {response.status_code}")
    print()
```

1.5 Fes una petició PATCH per modificar parcialment una publicació existent.

```python
url_patch = "https://jsonplaceholder.typicode.com/posts/1"

dato_patch = {
    "title": "Título modificado......"
}

respuesta = requests.patch(url_patch, json=dato_patch)

print(f"Estado Código: {respuesta.status_code}")
print(f"JSON: {respuesta.json()}")
```

## NIVEL 3

1. Utilitza l’API d’Open Data BCN per:

Cercar un dataset del teu interès mitjançant package_search i package_show. .



```python
URL_base = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/"

search_url = URL_base + "package_search"
params = {"q": "bicicleta"}

response = requests.get(search_url, params=params)

data = response.json()
results = data["result"]["results"]

for ds in results[:2]:
    print("Titulo:", ds["title"])
    print("ID:", ds["id"])
    print("\n" )
```

3.2 Dels resultats obtinguts, selecciona’n un que tingui recursos disponibles en CSV o JSON.


```python

dataset_id = "e616872e-0a5c-4128-be15-d2e49fbd78e4"  
show_url = URL_base + "package_show"
params = {"id": dataset_id}

response = requests.get(show_url, params=params)
dataset_info = response.json()["result"]

for resource in dataset_info["resources"]:
    print("Nombre:", resource["name"])
    print("Formato:", resource["format"])
    print("ID:", resource["id"])
    print("\n" )
```


3.3 Amb el resource_id del recurs seleccionat, realitza una consulta mitjançant datastore_search per:

Recuperar almenys 100 registres.


```python
resource_id = "66fa1382-476a-459b-b0c0-c666112d1843" 

datastore_url = URL_base + "datastore_search"
params = {"resource_id": resource_id, "limit": 100}

response = requests.get(datastore_url, params=params)
records = response.json()["result"]["records"]

print(f"Nombre de registres recuperats: {len(records)}")

for rec in records[:3]:
    print(rec)
```

3.4 Converteix els resultats a un DataFrame.



```python
df = pd.DataFrame(records)
df.head()
```
