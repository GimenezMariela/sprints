6.1 Crea una columna nova que sigui la data de naixament en format Datetime a partir de les columnes dia, mes i any. Utilitzant aquesta columna crea una funció que donada una data, et calculi l'edat actual a dia d'avui.


```python
df['fecha_nacimiento'] = pd.to_datetime(df.rename(columns={'Any': 'year', 'Mes': 'month', 'Dia': 'day'})[['year', 'month', 'day']])
df
```

6.2 Utilitza la funció que acabes de crear per generar una columna nova al DataFrame amb l'edat actual.

```python
def calcular_edad_actual(fecha_nacimiento):
    hoy = date.today()
    edad = hoy.year - fecha_nacimiento.year
    if (hoy.month, hoy.day) < (fecha_nacimiento.month, fecha_nacimiento.day):
        edad -= 1
    return edad

df['Edad_actual'] = df['fecha_nacimiento'].apply(calcular_edad_actual)
df
```

# 2 Ejercicio
2.1Utilitzant un bucle, exporta en 4 fitxers (format .xlsx o .csv) les dades de cada Grup Professional.

Per exemple: "dades_GrupA.xlsx" , "dades_GrupB.xlsx" ...

```python
grupos = df['Grup Professional'].unique()

for grupo in grupos:
    df_grupo = df[df['Grup Professional'] == grupo]
    filename = f"dades_{grupo.replace(' ','')}.xlsx" 
    df_grupo.to_excel(filename, index=False)
```

2.2 Exporta un 5è DataFrame en format .xlsx o .csv que contingui quants treballadors hi ha per cada Grup Professional, quin és el seu sou mig i quina és la seva edat mediana.

```python
resumen_grup = df.groupby('Grup Professional').agg(
    Treballadors=('Nom', 'count'),          # Número de trabajadores
    Sou_mig=('Salari mensual', 'mean'),     # Salario medio
    Edat_median=('Edad_actual', 'median')          # Edad mediana
).reset_index()

# Redondear valores
resumen_grup['Sou_mig'] = resumen_grup['Sou_mig'].round(2)
resumen_grup['Edat_median'] = resumen_grup['Edat_median'].astype(int)

# Exportar resumen
resumen_grup.to_excel("resumen_grups.xlsx", index=False)
```
